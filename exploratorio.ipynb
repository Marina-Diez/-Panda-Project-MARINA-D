```python
perc_null_row=attacks.isnull().sum(axis=1).apply(lambda x: x/attacks.shape[1]).sort_values(ascending=False)
perc_null_row
```

```python
# Only one colum is under 70% of missing data
cols = perc_null_col[perc_null_col < .7].index
cols
```
